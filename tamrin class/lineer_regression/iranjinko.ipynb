```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = ps.read_csv("D:/file madrese/maki sensei/lineer regression/iranjinko.csv")
data.head()
```

```python
year=data["Year"]
population=data["Population "]
#convert year to array
year_array=np.array(year).reshape(-1,1)
population_array=np.array(population)
plt.plot(year_array,population_array,"*")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(year_array,population_array)
```

```python
plt.plot(year_array,population_array,"*")
plt.plot(year_array,model.predict(year_array))
plt.show()
```

```python
result = model.predict([[2033]])
result
```
