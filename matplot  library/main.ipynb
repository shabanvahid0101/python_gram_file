```python
import matplotlib.pyplot as plt
```

# liner

```python
x=[1,2,3,4,5,6,7,8,9]
y=[2,4,3,7,1,8,9,11,4]
plt.plot(x,y,marker="*",color="green")
plt.title("plotchart")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

# barchart

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]
plt.title("barchart")
plt.xlabel("color")
plt.ylabel("count")
plt.bar(x,y,color="red")
plt.show()
```

# pie Chart

```python
x=["red","green","blue","orange"]
y=[5,19,2,4]
plt.pie(y,labels=x,autopct="%1.1f%%",colors=["red","green","blue","orange"])
plt.title("piechart")
plt.show()
```

# Histogram Chart

```python
import numpy as np
x=np.random.rand(1000,1)
plt.title("histogramchart")
plt.hist(x,bins=5)
plt.show()
```

```python
import matplotlib.pyplot as pl
import numpy as np
data = np.random.randn(1000)

pl.hist(data,bins=30,edgecolor="green",color="yellow")
pl.title("histogram")
pl.xlabel("value")
pl.ylabel("frequency")
pl.show()
```

# Hitmap

```python
x=np.random.rand(64,64)
plt.imshow(x,cmap="hot")
plt.title("hitmap")
plt.colorbar()
plt.show()
```

# hexabin plot

```python
x=np.random.randn(10000)
y=np.random.randn(10000)
#cmap = "Purples","Reds","Blues","magma"
plt.hexbin(x,y,gridsize=30,cmap='magma')
plt.colorbar(label='Density')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=["A","B","C"]
values_1=[10,15,12]
values_2=[8,17,14]
plt.plot(categories,values_1, label='2023',marker='o')
plt.plot(categories,values_2, label='2024',marker='o')
plt.xlabel("categories")
plt.ylabel("values")
plt.title("slopegraph")
plt.legend()
plt.show()
```

# draw a chessboard

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_chessboard():
    # Draw a chessboard
    chessboard=np.zeros((8,8))
    chessboard[1::2,::2]=1
    chessboard[::2,1::2]=1
    
    plt.figure(figsize=(4,4))
    plt.imshow(chessboard, cmap='gray',interpolation="nearest")
    plt.xticks([])
    plt.yticks([])
    plt.title('chessboard',fontsize=20)
    plt.show()
draw_chessboard()
```

# scatter plot

```python
riding=((17,18,21,22,19,21,25,22,25,24),(3,6,3.5,4,5,6.3,4.5,5,4.5,4))
swimming=((17,18,20,19,22,21,23,19,21,24),(8,9,7,10,7.5,9,8,7,8.5,9))
sailing=((31,28,29,36,27,32,34,35,33,39),(4,6.3,6,3,5,7.5,2,5,7,4))
plt.scatter(x=riding[0],y=riding[1],c='red',marker='s',label='riding')
plt.scatter(x=swimming[0],y=swimming[1],c='green',marker='o',label='swimming')
plt.scatter(x=sailing[0],y=sailing[1],c='blue',marker='*',label='sailing')

plt.xlabel('age')
plt.ylabel('hours')
plt.title('activities scatter plot')
plt.legend()

plt.show()
```
