
کاربرد دیپ لرنینگ
دیپ لرنینگ یک روش یادگیری ماشین است که برای پیش بینی و طبقه بندی داده
ها استفاده می شود. این روش بر اساس شبکه های عصبی مصنوعی است و می تواند
در بسیاری از زمینه ها از جمله پزشکی، مهندسی، اقتصاد و ... استفاده شود.

وقتی یک داده تو دفعات زیاد تغییر می کنه و تاثیر روی داده دیگری رو میشه با این پیش بینی کرد


```python
import tensorflow as tf
from tensorflow import keras
```

data definition

برای رگراسیون همیشه یک x , y داریم

برای اینکه مقدار x و y  رو مشخص کنیم یه رابطه لازم داریم

رابطه خطی
y = 3＊ｘー５


```python
x = tf.constant([-1,-3,5,7,9,0,8],dtype=float) 
y = x*3-5
```

```python
x.numpy()
```

```python
y.numpy()
```

model definition

Sequential کمک میکنه یک دنباله ای از لایه ها رو تعریف کنیم

Dense یعنی تک تک لایه ها با لایه های بعدی متصل بشن

units  تعداد نورون های یک لایه

input_shape لایه ورودی رو برای ما می سازه

در اینجا ما فقط با یک لایه و یک نورون مدل ساختیم چون رگرسیون خطی داریم

هرچی مدل ساده تر باشه روان تر میشه اجراش کرد اگر پیچیده بشه زمان بیشتری صرف یادگیری مدل میشه و ممکنه روی هر دستگاهی جواب نده

activation تعیین می کند که مدل ما از چه نوعی باشه اگر چیزی نزاریم یا none در نظر بگیریم خودش از نوع خطی میشناسه مدل رو



```python
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1],activation=None)])
```

در کراس برای اینکه مدل رو ترین کنیم باید قبلش کامپایل بشه 

optimizer
sgd سعی می کنه تابع هزینه رو بهینه کنه و وزن ها رو آپدیت کنه

loss=
mean_squared_error خطای مجموع مربعات 
(y-y_pred)^2 مقدار واقعی رو منحی مقدار پردیکت شده خطی رو بتوان 2 حساب میکنه




```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

میخوام ببینیم در مدل ما چه چیزایی داره

همیشه به تعداد ورودی ها به اضافه یک پارامتر داریم

```python
model.summary()
```

فیت کردن مدل

x 

y 

batch_size = 1
یعنی داده ها رو یکی یکی در نظر بگیر

اگر در کامپایل sqd تعریف شده باشه خودش این رو یک در نظر میگیره

epochs
یعنی تعداد دفعاتی که مدل رو روی داده ها تست می کنیم
اگر زیاد بزاریم ممکنه اور فیت بشه مدل



```python
model.fit(x,y,batch_size=1,epochs=500)
```

تست مدل 

x_test
یک بردار نمونه برای تست 
y_test
چیزی که واقعا برای ترین استفاده شده

```python
x_test=tf.constant([3,12,5,6,22,44,22,44,222,1,6],dtype=float)
y_test=x_test*3-5
```

```python
x_test.numpy()
```

```python
y_test.numpy()
```

تست مدل ساخته شده

```python
y_pred=model.predict(x_test)
y_pred
```

رسم نمودار

```python
import matplotlib.pyplot as plt
plt.scatter(x_test.numpy(),y_test.numpy(),color="red",linewidths=5)
plt.scatter(x_test.numpy(),y_pred,color="yellow",linewidths=2)
plt.show()
```
